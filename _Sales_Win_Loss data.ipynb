# PREPROCESSING OF DATA

# understanding LabelEncoder

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
#convert the categorical columns into string
encoded_value = label.fit_transform(["ram","shyam","sita","hari"])
print(encoded_value)
```

```python
print("Supplies Subgroup' : ",sales_data['Supplies Subgroup'].unique())
print("Region : ",sales_data['Region'].unique())
print("Route To Market : ",sales_data['Route To Market'].unique())
print("Opportunity Result : ",sales_data['Opportunity Result'].unique())
print("Competitor Type : ",sales_data['Competitor Type'].unique())
print("'Supplies Group : ",sales_data['Supplies Group'].unique())
```

```python
sales_data['Supplies Subgroup'] = label.fit_transform(sales_data['Supplies Subgroup'])
print(sales_data['Supplies Subgroup'])
```

```python
sales_data['Supplies Subgroup'] = label.fit_transform(sales_data['Supplies Subgroup'])
sales_data['Region'] = label.fit_transform(sales_data['Region'])
sales_data['Route To Market'] = label.fit_transform(sales_data['Route To Market'])
sales_data['Opportunity Result'] = label.fit_transform(sales_data['Opportunity Result'])
sales_data['Competitor Type'] = label.fit_transform(sales_data['Competitor Type'])
sales_data['Supplies Group'] = label.fit_transform(sales_data['Supplies Group'])
sales_data.head()
```

```python
cols=[col for col in sales_data.columns if col not in ['Opportunity Number', 'Opportunity Result']]
data = sales_data[cols]
target = sales_data['Opportunity Result']
data.head()
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(data,target,test_size = 0.30, random_state =10)
```

```python
print(xtrain)
```

```python
xtest
```

# MODEL BUILDING

Based on given dataset and checking by machine_learning_map of scikit-learn library,we can try out

Naive Bayes,
Linear SVC,
k-neighbours classifier

### NAIVE BAYES MODEL

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()
#train the data using training data and predict using test data
prediction = gnb.fit(xtrain,ytrain).predict(xtest)
print(prediction.tolist())
```

```python
accuracy_score(ytest, prediction, normalize = True)
```

### LinearSVC(linear support vector classification)

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

svc_model = LinearSVC(random_state =0)
prediction = svc_model.fit(xtrain,ytrain).predict(xtest)

print(prediction.tolist())
```

```python
accuracy_score(ytest, prediction, normalize = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier

n = KNeighborsClassifier(n_neighbors=3)
prediction = n.fit(xtrain,ytrain).predict(xtest)
print(prediction.tolist())
```

```python
accuracy_score(ytest,prediction)
```
