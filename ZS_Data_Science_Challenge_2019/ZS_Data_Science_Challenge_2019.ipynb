### dealing with missing values

```python
# filling missing values by linear interpolation
data['shot_id_number'].interpolate(method ='linear', limit_direction ='forward',inplace =True)
```

```python
data.head(15)
```

```python
data.isnull().any()
```

```python
# filling missing values for location_x and location_y with 0
data['location_x'].fillna(0,inplace =True)
data['location_y'].fillna(0,inplace =True)
#filling missing values of remaining minutes, remaining seconds and distance of shots with mean values
data['remaining_min'].fillna(data['remaining_min'].mean(),inplace = True)
data['remaining_sec'].fillna(data['remaining_min'].mean(),inplace = True)
data['distance_of_shot'].fillna(data['distance_of_shot'].mean(),inplace = True)

```

```python
#filling missing values in below columns with  previous values in the respective columns
data['game_season'].fillna(method = 'pad', inplace = True)
data['lat/lng'].fillna(method = 'pad', inplace = True)
data['power_of_shot'].fillna(method = 'pad', inplace = True)
data['knockout_match'].fillna(method = 'pad', inplace = True)
data['home/away'].fillna(method = 'pad', inplace = True)
```

```python
#filling missing values with self specified values
data['type_of_shot'].fillna('Not specified', inplace = True)
data['type_of_combined_shot'].fillna('Not specified', inplace = True)
```

```python
data['area_of_shot'].fillna('Not known', inplace = True)
data['shot_basics'].fillna('Not known', inplace = True)
data['range_of_shot'].fillna('Not known', inplace = True)
```

```python
#extracting the training dataset from whole dataset

train = data.where(data['is_goal'].isna() == False).dropna( how = 'all')
train.reset_index(inplace=True)
train.drop('index',1,inplace =True)
train.head(10)
```

```python
#extracting the test dataset from original dataset where value of is_goal is missing

test = data.where(data['is_goal'].isna()).dropna( how = 'all')
test.reset_index(inplace=True)
test.drop('index',1,inplace =True)
test.drop('is_goal',axis =1,inplace = True)

test.head()
```

```python
# dropping the lat/lng column

test.drop('lat/lng',axis =1,inplace =True)
train.drop('lat/lng',axis =1, inplace =True)
```

```python
#changing the datatype to float from string 

train['lat'] = train['lat'].astype(float)
train['lng'] = train['lng'].astype(float)
test['lat'] = test['lat'].astype(float)
test['lng'] = test['lng'].astype(float)
```

## preprocessing of data

```python
#Using label encoder for labelling data in different columns into categories
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label = LabelEncoder()
col =['game_season','area_of_shot','shot_basics','range_of_shot','home/away','type_of_shot','type_of_combined_shot']
for item in col:
    train[item] = label.fit_transform(train[item])
    test[item] = label.fit_transform(test[item])
```

```python
#Correlation plot for the feature correlation

import seaborn as sns


plt.style.use('fivethirtyeight')
sns.set_style("white")

corr = train.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
data, ax = plt.subplots(figsize=(16, 16))
plt.title('Ronaldo Feature Correlation')

# Generate a custom diverging colormap
cmap = sns.diverging_palette(260, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, vmax=1.2, square='square', cmap=cmap, mask=mask, 
            ax=ax,annot=True, fmt='.2g',linewidths=2)
```

```python
c_train = train[col]
c_train = pd.get_dummies(c_train,columns=col)
train = train.drop(col,axis =1)
train = pd.concat([train,c_train],axis =1)
```

```python
c_train.head()
```

```python
c_test = test[col]
c_test = pd.get_dummies(c_test,columns=col)
test = test.drop(col,axis =1)
test = pd.concat([test,c_test],axis =1)
```

```python
test.head()
```

```python
train.head()
```

```python
name_of_cols = train.columns.tolist()
name_of_cols1 =name_of_cols[:7] + name_of_cols[8:]

name_of_cols1
```

```python
## Scaling the training and test data

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = train[name_of_cols1]
y_train = train['is_goal']
X_test = test[name_of_cols1]
```

```python
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

## Model Building

```python
#Gradient Boosting Regression Model

from sklearn.ensemble import GradientBoostingRegressor
# get an instance of the model, using ‘ls’ : least squares regression as a loss function and the default learning rate which 0.1
gbr = GradientBoostingRegressor(loss ='ls', max_depth=6)
# fit the training data
gbr.fit (X_train, y_train)
# get the predicted values from the test set
y_pred = gbr.predict(X_test)
```
